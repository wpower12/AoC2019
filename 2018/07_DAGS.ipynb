```python
fn = "inputs/07_input.txt"
f  = open(fn, 'r')

preds = dict()
has_pred = dict()

def parse_line(line):
    raw_split = line.split()
    pred = raw_split[1]
    n    = raw_split[7]
    return n, pred

for line in f:
    n, pred = parse_line(line)
    
    if n not in preds:
        preds[n] = []
    
    if pred not in has_pred:
        has_pred[pred] = False
    
    preds[n].append(pred)
    has_pred[n] = True
    

initial_nodes = []
for n in has_pred:
    if not has_pred[n]:
        initial_nodes.append(n)
        if n in preds:
            del preds[n]
        print(n, "no preds")
    else:
        print(n, preds[n])

result = ""
stack = sorted(initial_nodes)
stack.reverse()

while len(stack) > 0:
    node = stack.pop()
    result += node
    if node in preds:
        del preds[node]
    
    new_items = []
    for p in preds:
        ps_preds = preds[p]
        if node in ps_preds:
            ps_preds.remove(node)
        if len(ps_preds) == 0:
            new_items.append(p)
    
    new_items.sort()
    new_items.reverse()
    for i in new_items:
        stack.append(i)
    
print(result)
    
    
```
